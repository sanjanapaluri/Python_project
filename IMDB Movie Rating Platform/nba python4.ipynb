*** Create a new column where height is in centimeters. Name the column as "Height (cm)"

Instruction: Height is given in feet and inches such as 6-2; 6 feet and 2 inches


Open Hint: 

Google the logic of converting height from feet - inches into centimeters
Use the string function to split the string
Understand and use the map function in Python
Create a dedicated function 
Use Pandas apply function with lambda to convert for all the data in the “Height” column

Note: Please make sure to get your code reviewed by mentors.

```python
df[['Feet', 'Inches']] = df['Height'].str.split('-', expand=True)
```

```python
df[['Feet', 'Inches']] = df[['Feet', 'Inches']].applymap(int)
```

```python
def convert_height(feet, inches):
    total_inches = feet * 12 + inches
    return round(total_inches * 2.54, 2)
```

```python
df['Height_cm'] = df.apply(lambda x: convert_height(x['Feet'], x['Inches']), axis=1)
```

*** Create a new column with the name “BMI Evaluation”. Assign the BMI status based on the following condition,


BMI <= 18.4 - "Underweight"

BMI <= 24.9 - "Healthy"

BMI <= 29.9 - "Over Weight"

BMI <= 34.9 - "Severely Over Weight"

BMI <= 39.9 - "Obese"
39.9 - "Severely Obese."

```python
conditions = [
    (df['BMI'] <= 18.4),
    (df['BMI'] <= 24.9),
    (df['BMI'] <= 29.9),
    (df['BMI'] <= 34.9),
    (df['BMI'] <= 39.9),
    (df['BMI'] > 39.9)
]
```

```python
bmi_categories = [
    'Underweight',
    'Healthy',
    'Over Weight',
    'Severely Over Weight',
    'Obese',
    'Severely Obese'
]
```

```python
df["BMI Evaluation"] = np.select(conditions, bmi_categories)
```

*** Compute the mean salary of the team using group by and create dataframe df_1.

Join the tables df_1 and df to add the mean salary of each team in every row [Similar to Avg Over Partition By Windows Function in SQL]

```python
df_1 = df.merge(df.groupby("Team")["Salary"].mean(),on="Team",suffixes=("_player","_team"))
df_1
```
